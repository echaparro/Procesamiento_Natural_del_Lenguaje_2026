#### 8. Aplica un pipeline de limpieza de texto sobre la siguiente reseña usando `re.sub`. Los pasos deben ser en orden:
1. Reemplazar URLs por el token `[URL]`
2. Reemplazar correos electrónicos por el token `[EMAIL]`
3. Reemplazar números de teléfono (formato `(XX) XXXX-XXXX`) por `[TELÉFONO]`
4. Eliminar caracteres no ASCII (emojis, símbolos especiales)
5. Normalizar espacios múltiples a uno solo
6. Convertir a minúsculas

```
reseña = """¡¡INCREÍBLE producto!! 😍🔥 Cómpralo en https://tienda.com/promo?ref=123
Contacto: soporte@tienda.com o llama al (55) 8765-4321
   Mucho   espacio   extra   aquí   """
```

````python
import re

reseña = """¡¡INCREÍBLE producto!! 😍🔥 Cómpralo en https://tienda.com/promo?ref=123
Contacto: soporte@tienda.com o llama al (55) 8765-4321
   Mucho   espacio   extra   aquí   """

print(reseña)
print("="*25,"1.Reemplazar URLs por [URL]","="*25)
reseña = re.sub(r'https?://[^\s]+', '[URL]', reseña)
print(reseña)
print()

print("="*25,"2.Reemplazar correos por [EMAIL]","="*25)
reseña = re.sub(r'\b[\w\.-]+@[\w\.-]+\.\w+\b', '[EMAIL]', reseña)
print(reseña)
print()

print("="*25,"3.Reemplazar números de teléfono (formato (XX) XXXX-XXXX) por [TELÉFONO]","="*25)
reseña = re.sub(r'\(\d{2}\)\s?\d{4}-\d{4}', '[TELÉFONO]', reseña)
print(reseña)
print()

print("="*25,"4.Eliminar caracteres no ASCII (emojis, símbolos especiales)","="*25)
reseña = re.sub(r'[^\x00-\x7F]+', '', reseña)
print(reseña)
print()

print("="*25,"5.Normalizar espacios múltiples a uno solo","="*25)
reseña = re.sub(r'\s+', ' ', reseña)
print(reseña)  
print()

print("="*25,"6.Convertir a minúsculas","="*25)
reseña = reseña.lower()
print(reseña)
print()

````

## Módulo `string`

#### 9. Usa el módulo `string` para limpiar la puntuación del siguiente texto. Realiza los siguientes pasos:
1. Elimina toda la puntuación (incluye `¡` y `¿`) usando `str.maketrans` y `.translate()`
2. Convierte el resultado a minúsculas
3. Divide el texto en una lista de palabras (tokens) usando `.split()`
4. Imprime: el texto limpio, la lista de tokens y el número total de tokens

```
texto = "¡Hola, mundo! ¿Cómo estás? El NLP... es fascinante."
```

````python
import string

texto = "¡Hola, mundo! ¿Cómo estás? El NLP... es fascinante."

# Paso 1: Eliminar puntuación con str.maketrans y .translate()
#print(string.punctuation) #!"#$%&'()*+,-./:;<=>?@[\]^_`{|}~
# #No trae todos los signos de puntuación, se tiene que agregar los faltantes
puntuacion_extra = "¡¿"
puntuacion_completa = string.punctuation + puntuacion_extra
tabla_puntuacion = str.maketrans('', '', puntuacion_completa)
texto_sin_puntuacion = texto.translate(tabla_puntuacion)

# Paso 2: Convertir a minúsculas
texto_minusculas = texto_sin_puntuacion.lower()

# Paso 3: Dividir en tokens
tokens = texto_minusculas.split()

# Paso 4: Imprimir resultados
print("="*25,"Resultados","="*25)
print("Texto original:", texto)
print("Texto sin puntuación:", texto_sin_puntuacion)
print("Texto en minúsculas:", texto_minusculas)
print("Tokens:", tokens)
````

## Conceptos Fundamentales de NLP

#### 10. Dado el siguiente mini-corpus, responde las preguntas **con código**:

```
corpus = [
    "el gato duerme en la alfombra",
    "el perro corre en el parque",
    "la gata y el perro son amigos",
    "el parque tiene una alfombra verde",
]
```

a) Calcula el **vocabulario** del corpus (tokens únicos en minúsculas, sin puntuación). ¿Cuántos tipos (`types`) hay?

b) ¿Cuántos tokens en total hay en todo el corpus?

c) Calcula la **frecuencia** de cada tipo. ¿Cuáles son las 5 palabras más frecuentes? ¿Se observa la **Ley de Zipf**?

d) ¿Qué palabras del vocabulario podrían considerarse **stopwords**? Identifícalas filtrando con NLTK y muestra el vocabulario resultante sin ellas.

````python
from collections import Counter
from nltk.corpus import stopwords

corpus = [
    "el gato duerme en la alfombra",
    "el perro corre en el parque",
    "la gata y el perro son amigos",
    "el parque tiene una alfombra verde",
]

# a) Vocabulario: tokens únicos
vocabulario = set()
tokens_totales = []
for documento in corpus:
    tokens = documento.split()
    vocabulario.update(tokens)
    tokens_totales.extend(tokens)
print("="*25,"Vocabulario","="*25)
print(vocabulario)
print()

# b) Total de tokens
total_tokens = len(vocabulario)
print("="*25,"Total de Tokens","="*25)
print(total_tokens)

````

````python
# c) Frecuencias — Ley de Zipf
frecuencias = Counter(tokens_totales)
print("="*25,"Frecuencia 5 palabras más frecuentes","="*25)
for i, (palabra, freq) in enumerate(frecuencias.most_common(), 1):
    print(f"{palabra}: {freq}")
    if i==5:
        break

print()

````

#### La Ley de Zipf

En cualquier idioma natural, pocas palabras son muy frecuentes y la mayoría son raras. Esto es la **Ley de Zipf**:

$$f(r) \propto \frac{1}{r}$$

donde $f$ es la frecuencia y $r$ es el rango (posición en el ranking).

````python
print("="*25,"Ley de Zipf","="*25)
print("="*5,"No se cumple la ley de Zipf, los rangos de frecuencias varian mucho")
print("="*5,"La cantidad de palabras únicas es muy pequeña para poder observar la ley de Zipf")
print("="*5,"f × r no es constante, deberia ser mas constante, no exacto pero si cercano ")
print("Rango | Palabra    | Frecuencia | f × r")
print("-" * 45)
for i, (palabra, freq) in enumerate(frecuencias.most_common(), 1):
    producto = freq * i
    print(f"{i:4}  | {palabra:10} | {freq:9} | {producto:6.1f}")

````

````python
# d) Vocabulario sin stopwords
stop_words = set(stopwords.words('spanish')) #stopwords en español
vocabulario_filtrado = [palabra for palabra in vocabulario if palabra not in stop_words]
stop_words_filtrados = [palabra for palabra in vocabulario if palabra in stop_words]
print("="*25,"Stopwords encontrados","="*25)
print(stop_words_filtrados)
print()
print("="*25,"Vocabulario sin stopwords","="*25)
print(vocabulario_filtrado)
print()
total_tokens_filtrado = len(vocabulario_filtrado)
print("="*25,"Total de Tokens (sin stopwords)","="*25)
print(total_tokens_filtrado)
print()
````
